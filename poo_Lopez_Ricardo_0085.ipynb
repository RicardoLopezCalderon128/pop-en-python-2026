# lenguaje python crear clase "animal " con los atributos "nombre" raza crear el conructor animal y las funciones ladrar (), correr (),crear los objetos pitbull,chihuahua nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor __init__ asigna los valores a los atributos
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} ({self.raza}) está ladrando: ¡Guau! 😆")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo por el parque. ✨")
    def Morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 😜")
# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto pitbull con datos específicos
pitbull = Animal("Thor", "Pitbull")

# Creamos el objeto chihuahua con datos específicos
chihuahua = Animal("Pipo", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "Callejero")
# --- Ejecución de las funciones ---

# Acciones del Pitbull
print("--- Datos del primer perro ---")
pitbull.ladrar()
pitbull.correr()

print("\n--- Datos del segundo perro ---")

# Acciones del Chihuahua
chihuahua.ladrar()
chihuahua.correr()

print("\n--- Datos del tercer perro ---")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()


print("programa creado por Lopez Ricardo NL = 29")
```
